```python
import matplotlib.pyplot as plt
```

```python
xs = []
ys = []
with open('input.txt') as f:
    for line in f:
        x, y = map(int, line.strip().split(','))
        xs.append(x)
        ys.append(y)
```

```python
import matplotlib.patches as patches

# Sample result points - to be replaced with the solution
p1 = (1000, 2000)
p2 = (10000, 10000)

square = patches.Rectangle((p1[0], p2[1]), p2[0] - p1[0], p1[1] - p2[1])
fig, ax = plt.subplots()
ax.add_patch(square)

plt.plot(xs, ys)
```
